## OpenCV Camera Testing

```python
# https://stackoverflow.com/questions/57577445/list-available-cameras-opencv-python

def list_ports():
    """
    Tests the ports and returns a tuple with the available ports and the ones that are working.
    """
    non_working_ports = []
    dev_port = 0
    working_ports = []
    available_ports = []
    while len(non_working_ports) < 6: # if there are more than 5 non working ports stop the testing. 
        camera = cv2.VideoCapture(dev_port)
        if not camera.isOpened():
            non_working_ports.append(dev_port)
            print("Port {} is not working.".format(dev_port))
        else:
            is_reading, img = camera.read()
            w = camera.get(3)
            h = camera.get(4)
            if is_reading:
                print("Port {0} is working and reads images ({1} x {2})" .format(dev_port,h,w))
                working_ports.append(dev_port)
            else:
                print("Port {0} for camera ( {1} x {2}) is present but does not reads.".format(dev_port,h,w))
                available_ports.append(dev_port)
        dev_port +=1
    return available_ports,working_ports,non_working_ports
```

### Model

```python
# load the model
model = models.resnet50(weights='DEFAULT')

print(model)
model_weights = [] # we will save the conv layer weights in this list
conv_layers = [] # we will save the 49 conv layers in this list
# get all the model children as list
model_children = list(model.children())
```

```python
type(model)
```

```python
# counter to keep count of the conv layers
counter = 0 
# append all the conv layers and their respective weights to the list
for i in range(len(model_children)):
    if type(model_children[i]) == nn.Conv2d:
        counter += 1
        model_weights.append(model_children[i].weight)
        conv_layers.append(model_children[i])
    elif type(model_children[i]) == nn.Sequential:
        for j in range(len(model_children[i])):
            for child in model_children[i][j].children():
                if type(child) == nn.Conv2d:
                    counter += 1
                    model_weights.append(child.weight)
                    conv_layers.append(child)
print(f"Total convolutional layers: {counter}")
```

```python
# take a look at the conv layers and the respective weights
for weight, conv in zip(model_weights, conv_layers):
    # print(f"WEIGHT: {weight} \nSHAPE: {weight.shape}")
    print(f"CONV: {conv} ====> SHAPE: {weight.shape}")
```

```python
# %%timeit -n 30
# different approach (negative values to zero, then mapping
# to pyplot (mapping minimum value as 0 and maximum as 1)
# pyplot method preferred
custom_color_adaption = False
norm = matplotlib.colors.Normalize() # use Matplotlib normalization
n_layer = 104
#layers = [0,1,2,3,4,20,30,60,80,90,100,104]
layers = [0,1,2,30,60,90,100,104]
use_plt = False
n_img = 8 # grid size

# iterate through chosen layers
for layer_num in layers:

    images = model.tensor_collection[layer_num][0][0] # retrieve images from custom dictionary attribute
    
    # generate plt figure
    if use_plt:
        fig = plt.figure(num=layer_num, figsize=(30, 30))

    for i in range(len(images)):
        # set maximum retrieval number to n_img^2 images, even if layer size is greater than this number
        if (i+1) > (n_img**2):
            break
        
        img_layer = images[i].detach().numpy() # retrieve image as NumPy array
        img_layer = norm(img_layer) * 255 # normalize values to range [0,1] and multiply by max value 255 (color model)
        img_layer = img_layer.round(decimals=0).astype(np.uint8) # round and cast as int object

        # testing purposes, different normalization method than Matplotlib
        if custom_color_adaption:
            img_layer = np.where(img_layer < 0, 0, img_layer)
            img_layer = img_layer / img_layer.max() * 255
        
        # concatenate the max. n_img arrays as n_img^2 block
        if i == 0:
            grid_list = []
        
        if (i % n_img) == 0: # define row size (e.g. n_img=8 --> row length is 8)
            row_list = [] # (re-)initialize list of rows
        
        row_list.append(img_layer) # add image to row
        
        if len(row_list) == n_img:
            grid_list.append(row_list) # append complete rows to collection list
        
        #print(f"size row list: {len(row_list)}")
        # add img_layer to subplot of plt figure
        if use_plt:
            fig.add_subplot(8, 8, i + 1)
            plt.imshow(img_layer, cmap='gray')
            plt.axis("off")
    
    img_block = np.block(grid_list)
    
    # saving plt figure
    if use_plt:
        print(f"Saving  feature maps...")
        plt.savefig(f"output/FeatureMap/Yolo_layer_{layer}.png")
        #plt.show()
        plt.close()
    
    
    is_percentage_resize = False
    if is_percentage_resize:
        if img_block.shape[1] > 1200:
            scale_percent = 40 # percent of original size
            width = int(img_block.shape[1] * scale_percent / 100)
            height = int(img_block.shape[0] * scale_percent / 100)
            dim = (width, height)
            img_block = cv2.resize(img_block, dim, interpolation = cv2.INTER_AREA)
        elif img_block.shape[1] < 800:
            scale_percent = 160 # percent of original size
            width = int(img_block.shape[1] * scale_percent / 100)
            height = int(img_block.shape[0] * scale_percent / 100)
            dim = (width, height)
            img_block = cv2.resize(img_block, dim, interpolation = cv2.INTER_CUBIC)
        
    else:
        target_width = 800
        scale_percent = float(target_width / img_block.shape[1])
        height = int(img_block.shape[0] * scale_percent)
        dim = (target_width, height)
        if scale_percent > 1.0: # upsampling
            img_block = cv2.resize(img_block, dim, interpolation = cv2.INTER_CUBIC)
        else: # downsampling
            img_block = cv2.resize(img_block, dim, interpolation = cv2.INTER_AREA)
    
    cv2.imshow(f"Layer {layer_num}", img_block)
    #cv2.imshow("Test", data)
    cv2.waitKey(0)
    
cv2.destroyAllWindows()
    
    #cv2.imshow("Display window", img)
    #k = cv.waitKey(0)
    
    

#cv2.destroyAllWindows()
```
